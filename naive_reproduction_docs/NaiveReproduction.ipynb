Assumtions:
Based on the content in the paper I have made the assumtions of the following hypotheses.
H0-1: There will be no significant difference in the number of trials.
H1-1: There will be a significant difference in the number of trials.
H0-2: There will be no significant difference in the number time taken.
H1-2: There will be a significant difference in the number time taken.
H1-3: There will be a significant difference in the distance error.
H0-3: There will be no significant difference in the distance error.
H1-3: There will be a significant difference in the distance error.

Note:
Distance refers to mm,
Time refers to seconds

```python
# software & usability effectiveness calculations (mean, median, SD, IQR, SE)
effectiveness_mean = qdf['Usability & Effectiveness'].mean()
effectiveness_median = qdf['Usability & Effectiveness'].median()
effectiveness_sd = qdf['Usability & Effectiveness'].std()
effectiveness_iqr = qdf['Usability & Effectiveness'].quantile(0.75) - qdf['Usability & Effectiveness'].quantile(0.25)
effectiveness_se = qdf['Usability & Effectiveness'].sem()

# Graphic User Interface calculations (mean, median, SD, IQR, SE)
gui_mean = qdf['Graphic User Interface'].mean()
gui_median = qdf['Graphic User Interface'].median()
gui_sd = qdf['Graphic User Interface'].std()
gui_iqr = qdf['Graphic User Interface'].quantile(0.75) - qdf['Graphic User Interface'].quantile(0.25)
gui_se = qdf['Graphic User Interface'].sem()

# User Support calculations (mean, median, SD, IQR, SE)
support_mean = qdf['User Support'].mean()
support_median = qdf['User Support'].median()
support_sd = qdf['User Support'].std()
support_iqr = qdf['User Support'].quantile(0.75) - qdf['User Support'].quantile(0.25)
support_se = qdf['User Support'].sem()

# Learning Time calculations (mean, median, SD, IQR, SE)
learning_mean = qdf['Learning Time'].mean()
learning_median = qdf['Learning Time'].median()
learning_sd = qdf['Learning Time'].std()
learning_iqr = qdf['Learning Time'].quantile(0.75) - qdf['Learning Time'].quantile(0.25)
learning_se = qdf['Learning Time'].sem()

# user flexibility calculations (mean, median, SD, IQR, SE)
flexibility_mean = qdf['User Flexibility'].mean()
flexibility_median = qdf['User Flexibility'].median()
flexibility_sd = qdf['User Flexibility'].std()
flexibility_iqr = qdf['User Flexibility'].quantile(0.75) - qdf['User Flexibility'].quantile(0.25)
flexibility_se = qdf['User Flexibility'].sem()

# Troubleshooting calculations (mean, median, SD, IQR, SE)
troubleshooting_mean = qdf['Troubleshooting'].mean()
troubleshooting_median = qdf['Troubleshooting'].median()
troubleshooting_sd = qdf['Troubleshooting'].std()
troubleshooting_iqr = qdf['Troubleshooting'].quantile(0.75) - qdf['Troubleshooting'].quantile(0.25)
troubleshooting_se = qdf['Troubleshooting'].sem()

# Error Handling calculations (mean, median, SD, IQR, SE)
error_handling_mean = qdf['Error Handling'].mean()
error_handling_median = qdf['Error Handling'].median()
error_handling_sd = qdf['Error Handling'].std()
error_handling_iqr = qdf['Error Handling'].quantile(0.75) - qdf['Error Handling'].quantile(0.25)
error_handling_se = qdf['Error Handling'].sem()

# Adaptibility calculations (mean, median, SD, IQR, SE)
adaptibility_mean = qdf['Adaptibility'].mean()
adaptibility_median = qdf['Adaptibility'].median()
adaptibility_sd = qdf['Adaptibility'].std()
adaptibility_iqr = qdf['Adaptibility'].quantile(0.75) - qdf['Adaptibility'].quantile(0.25)
adaptibility_se = qdf['Adaptibility'].sem()

```

```python
# table showing results of stats overview for the questionnaire data
summary_questionnaire_data = {'Questionnaire': ['Usability & Effectiveness', 'Graphic User Interface', 'User Support','Learning Time','User Flexibility','Troubleshooting', 'Error Handling', 'Adaptibility' ],
                              'Mean': [effectiveness_mean, support_mean, flexibility_mean, learning_mean, flexibility_mean, troubleshooting_mean, error_handling_mean, adaptibility_mean],
                              'Median':[effectiveness_median, support_median, flexibility_median, learning_median, flexibility_median, troubleshooting_median, error_handling_median, adaptibility_median],
                              'SE': [effectiveness_se, support_se, flexibility_se, learning_se, flexibility_se, troubleshooting_se, error_handling_se, adaptibility_se],
                              'SD': [effectiveness_sd, support_sd, flexibility_sd, learning_sd, flexibility_sd, troubleshooting_sd, error_handling_sd, adaptibility_sd],
                              'IQR':[effectiveness_iqr, support_iqr, flexibility_iqr, learning_iqr, flexibility_iqr, troubleshooting_iqr, error_handling_iqr, adaptibility_iqr]

                            }

summary_questionnaire_qdf = pd.DataFrame(summary_questionnaire_data).round(1)
summary_questionnaire_transposed_qdf = summary_questionnaire_qdf.set_index('Questionnaire').transpose().round(2)
summary_questionnaire_transposed_qdf.index.name = None
summary_questionnaire_final_df = summary_questionnaire_transposed_qdf.transpose()

display(summary_questionnaire_final_df)
```
